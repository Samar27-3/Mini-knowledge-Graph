## Step 2: Define Our Mini Knowledge Graph
We'll create a list of triples: (subject, relationship, object)

88f4f5c75d9608b7e20ccd3f28797833.jpg

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define some simple knowledge triples
triples = [
    ("Alice", "has_disease", "Flu"),
    ("Alice", "has_symptom", "Fever"),
    ("Bob", "has_disease", "COVID-19"),
    ("Bob", "has_symptom", "Cough"),
    ("Flu", "has_symptom", "Fever"),
    ("COVID-19", "has_symptom", "Loss of Smell")
]
```

##Step 3: Build the Graph

images (2).jpeg

```python
# Create a directed graph
G = nx.DiGraph()

# Add edges from the triples
for subj, rel, obj in triples:
    G.add_edge(subj, obj, label=rel)
```

## Step 4: Visualize the Knowledge Graph

```python
# Draw the graph
import networkx as nx
import matplotlib.pyplot as plt
#Create a directed graph
G = nx.DiGraph()
# Define some simple knowledge triples
triples = [
    ("Ali","has_disease","cancer"),
    ("Ali","has_symptom","pain"),
    ("Samir","has_major","computer science"),
    ("Samir","has_graduated","LU university"),
    ("cancer","has_symptom","pain"),
    ("LU university","has_major","computer science"),
]
# Add edges from the triples
for subj, rel, obj in triples:
  G.add_edge (subj, obj, label=rel)
# Define manual positions python
pos = {
    "Ali": (2,2),
    "cancer" : (1,2),
    "pain" : (2,1),
    "Samir" : (2,0),
    "computer science" : (0,2),
    "LU university" : (2,0.5),

}



plt.figure(figsize=(10, 6))
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos, with_labels=True, node_color="lightgreen", node_size=2300, font_size=6, edge_color="red")
edge_labels = nx.get_edge_attributes(G, "label")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="black")
plt.title("Mini knowledge Graph ")

# Save the figure before showing it
plt.savefig('knowledge_graph.png')

plt.show()
```
